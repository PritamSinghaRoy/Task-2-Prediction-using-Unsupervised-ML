# Task 2 Prediction using Unsupervised ML

      From the given ‘Iris’ dataset, predict the optimum number of clusters and
      represent it visually.
      
      Dataset : https://bit.ly/3kXTdox


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the data
data=pd.read_csv("Iris.csv")
data.head()
```

```python
data=data.drop("Id",axis=1)
```

```python
data.isna().sum()
```

# Data Visualization

```python
sns.set_style("darkgrid")
sns.pairplot(data,hue="Species",palette="hls")
```

```python
x=data.drop(["Species"],axis=1)
x.head()
```

# Finding optimum numbers of clusters for K Means

```python
from sklearn.cluster import KMeans
x=data.iloc[:,[0,1,2,3]].values
wcss=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    print("k:",i,"wcss: ",kmeans.inertia_)
plt.plot(range(1,10),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()

```

## Let's take 5 as no of clusters





# Applying KMeans to the dataset

```python
kmeans=KMeans(n_clusters=3,init="k-means++",max_iter=300,n_init=10,random_state=0)
y_kmeans=kmeans.fit_predict(x)   
   
```

## Predicted values

```python
y_kmeans
```

## Plotting the cluster on the first two columns

```python
plt.figure(figsize=(12,12))
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=80,c="blue",label="Iris-setosa")
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=80,c="green",label="Iris-versicolour")
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=80,c="orange",label="Iris-virginica")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c="yellow",label="Centroids")
plt.legend()
```
